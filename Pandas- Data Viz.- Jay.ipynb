# A) Plot in Pandas using Seaborn
* SYNTAX: df_name['ref_col'].plot/ plot_type / plot.type(sns_arg_in_syntax)
    * sns_arg_in_syntax =kind, bins, kde, annnot......etc
    
## Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)
___

# B) >> 3-D Plotting (IMP)

* Based on color density
* (SYNTAX): df_name.plot.type(x='col1', y='col2', c='ref_col3')
    * c = color density based on 3rd col.

```python
# Based on color density(SYNTAX)
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm') #VIMP------------------------------->>>>>
                                    # Shows Graph of (A, B) with color density of col. (C) --> 3D plot
```

* SYNTAX(Based on Size, in comparision): df_name.plot.type(x='col1',y='col2',s=df_name['ref_col3'])
    * s= size based on 3rd col. in comparision to A & B

```python
#Based on size comparision (Comparitively)
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

# C) Hexbin in Pandas
