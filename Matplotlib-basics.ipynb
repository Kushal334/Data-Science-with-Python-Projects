#### B)Figure Size, Aspect Ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:
    
- NOTE1: Using figure() || Requiers Manual Add axes || 'for' for iteration
- NOTE2 : Using subplots() || fig,axes || Requires I/P col & rows no. || Indexing for iteration

```python
#1 Using figure() || Requiers Manual Add axes || 'for' for iteration

fig = plt.figure(figsize = (8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#2 Using subplots() || fig,axes || Requires I/P col & rows no. || Indexing for iteration
fig,axes = plt.subplots(nrows = 2, ncols=1, figsize = (8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### A) Plot Appearance

#### 1.  Color | linewidth | alpha | linestyle | marker | markersize | markerfacecolor | markeredgewidth | markeredgecolor

* SYNTAX: plt_name.plot(x,y,color = 'color_code|name', linewidth or lw = value, alpha = value, linestyle or ls = 'pattern or pattern name', marker = 'mark_pattern', markersize = value)
    
* `linewidth` or `lw`    to decide width of line
* `alpha`                controls line transperancy
* `linestyle` or `ls`     to decide the style of line
* `marker`                used to mark the points on the line
* `markersize`            used to decide size of marker
* `markerfacecolor`
* `markeredgewidth`
* `markeredgecolor`
Can use RGB Hex codes as well

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color = 'orange',lw = 3,alpha=0.5,linestyle = 'dashdot')
axes.plot(y,x,color = 'purple', lw = 4, ls = 'solid',marker = 'o',markersize = 10,
          markerfacecolor = 'y',markeredgewidth = 2, markeredgecolor = 'blue')
```

### B) LINE and MARKER styles

* SYNTAX(Custom_dash): line, =plt_name.plot(x,y,color=,lw=)
                       line.set_dashes([line_length, space_lenght, ....])

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

## C) Plot Range

* We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:
* SYNTAX(Range Limit): plt_name.set_xlim([,])    
* SYNTAX(tight axis): plt_name[index].axis('tight')

```python
fig,axes = plt.subplots(1,3,figsize =(12,4))

axes[0].plot(x,x**2,x,x**3) #2 attributes for line-1 & 2 attributes for line-2
axes[0].set_title("Default axes range")

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title("Tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range")
```
